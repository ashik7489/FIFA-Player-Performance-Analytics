```python
def height_to_cm(x):
    if pd.isna(x):
        return None
    try:
        feet, inches = x.replace('"','').split("'")
        return int(feet) * 30.48 + int(inches) * 2.54
    except:
        return None
```

```python
df['height_cm'] = df['height'].apply(height_to_cm)
```

```python
df['height_cm'] = df['height_cm'].fillna(df['height_cm'].median())
```

```python
df.drop(columns=['height'], inplace=True)
```

```python
df['weight'] = (
    df['weight']
    .str.replace('lbs','', regex=False)
    .astype(float)
)

df['weight'] = df['weight'].fillna(df['weight'].median())
```

```python
import numpy as np

def convert_money(x):
    if pd.isna(x):
        return np.nan
    x = str(x).replace('€','')
    if 'M' in x:
        return float(x.replace('M','')) * 1_000_000
    if 'K' in x:
        return float(x.replace('K','')) * 1_000
    return float(x)

df['value'] = df['value'].apply(convert_money)
df['wage'] = df['wage'].apply(convert_money)
df['release_clause'] = df['release_clause'].apply(convert_money)
```

```python

def convert_money(x):
    if pd.isna(x):
        return np.nan

    x = str(x).strip()

    # handle text / invalid cases FIRST
    if x in ['Not Available', 'NA', 'N/A', 'None', '', '-', '--']:
        return np.nan

    x = x.replace('€','')

    try:
        if 'M' in x:
            return float(x.replace('M','')) * 1_000_000
        elif 'K' in x:
            return float(x.replace('K','')) * 1_000
        else:
            return float(x)
    except:
        return np.nan



```

```python
df['release_clause_eur'] = df['release_clause '].apply(convert_money)
```
